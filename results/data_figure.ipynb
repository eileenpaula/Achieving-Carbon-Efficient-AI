```python
from datasets import load_dataset
import matplotlib.pyplot as plt
import pandas as pd

# Load the Amazon Polarity dataset
dataset = load_dataset("amazon_polarity")

# Split off a 75% sample of the training set
train_subset = dataset["train"].train_test_split(test_size=0.25)["train"]
test_dataset = dataset["test"]  # Keep the full test set for evaluation
```

```python
# Set font and general style
plt.rcParams["font.family"] = "serif"  # Use Times New Roman or similar
plt.rcParams["axes.labelsize"] = 16  # Increase axis label font size
plt.rcParams["xtick.labelsize"] = 14  # Increase x-tick font size
plt.rcParams["ytick.labelsize"] = 14  # Increase y-tick font size
plt.rcParams["legend.fontsize"] = 14  # Increase legend font size
plt.rcParams["axes.titlesize"] = 18  # Increase title font size

# Count label distribution in the training subset
label_counts = train_subset["label"]
label_distribution = pd.Series(label_counts).value_counts()

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(label_distribution, labels=["Positive", "Negative"], autopct='%1.1f%%', startangle=140)

plt.show()
```

```python
# Sizes for training subset and full test dataset
train_size = len(train_subset)
test_size = len(test_dataset)

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie([train_size, test_size], labels=["Training Subset", "Testing Set"], autopct='%1.1f%%', startangle=140)

plt.show()
```
